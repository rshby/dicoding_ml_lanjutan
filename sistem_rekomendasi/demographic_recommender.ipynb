```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(context="talk")
```

```python
df = (pd.read_csv("dataset/IMDb movies.csv").loc[:, "original_title":"votes"].drop(["language", "production_company", "description"], axis="columns"))
df["year"] = df["year"].str.replace("TV Movie 2019", "2019")
df = df.astype({"avg_vote":"float", "votes":"float", "duration":"float", "year":"float"})
df.dropna(subset=["year"], inplace=True)
```

## Step 1: Filtering

```python
genre = ["Horror", "Drama"]
year = [2018, 2020]
duration = [90, 150]
top_k = 10

hasil_perdiksi = df[(df["year"].between(year[0], year[1])) & (df["duration"].between(duration[0], duration[1])) & (df["genre"].str.contains("Drama")) & (df["genre"].str.contains("Crime"))]
```

# Improve Demographic Recomendation System
bikin scoring manual dengan formula sendiri

```python
m = df["votes"].quantile(0.6)
C = (df["avg_vote"] * df["votes"])/(df["votes"].sum())
df = df.loc[df["votes"] > m]
df["socer"] = (df["avg_vote"] * df["votes"] + (C + m)) / (df["votes"] + m)
```

## Coba Prediksi dengan beberapa Filter

```python
df.loc[(df["genre"].str.contains("Horror")) & (df["genre"].str.contains("Comedy")) == True].sort_values(by="socer", ascending=False)
```
